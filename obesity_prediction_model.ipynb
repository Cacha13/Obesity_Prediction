```python
# Checking the porpotions for the categorical columns:
cat_columns = ['Gender', 'family_history', 'FAVC', 'SMOKE', 'SCC', 'MTRANS', 'CALC', 'CAEC', 'Obesity']
for i in cat_columns:
    print(data[i].value_counts(normalize = True))
    print('')
```

```python
# Checking the outliers from the numeric columns and plotting the distributions and the boxplots:

numeric_data = data.select_dtypes(include=['number'])

for i in numeric_data:
    print(str.upper(i) + ':')
    Q1 = numeric_data[i].quantile(0.25)
    Q3 = numeric_data[i].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((numeric_data[i] < (Q1 - 1.5 * IQR)) | (numeric_data[i] > (Q3 + 1.5 * IQR))).sum()
    print( 'Nº of outliers: ' + str(outliers))
    plt.figure(figsize = (10,6))
    fig, axes = plt.subplots(1,2, figsize = (12,5))
    axes[0].hist(numeric_data[i])
    axes[1].boxplot(numeric_data[i])
    fig.suptitle(i.upper(), fontsize = 16)
    plt.show()
    plt.clf()
```

DATA CLEANING

```python
# Removing outliers rows from the Height and Weight columns:
outliers_columns = ['Height', 'Weight']
for x in outliers_columns:
    q1 = data[x].quantile(0.25)
    q3 = data[x].quantile(0.75)
    iqr = q3 - q1
    data = data[(data[x] >= (q1 - 1.5 * iqr)) & (data[x] <= (q3 + 1.5 * iqr))]
```

```python
# Changing the values from the categorical columns to be numerical values:
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

# To encode the binary columns we use the LabelEncoder being in the Gender (Female:0, Male:1) and the rest of the columns (No:0, Yes:1) and we will also use in the column MTRANS because the order isn't relevant

for i in cat_columns[:6]:
    data[i] = encoder.fit_transform(data[i])
    print(data[i].value_counts(normalize = True))

#  For the others we will use the .map() function:
for i in cat_columns[6:8]:
    data[i] = data[i].map({'no': 0, 
                           'Sometimes': 1, 
                           'Frequently': 2, 
                           'Always': 3})
    print(data[i].value_counts(normalize = True))

data['Obesity'] =  data['Obesity'].map({'Insufficient_Weight': 0, 
                                        'Normal_Weight': 1, 
                                        'Overweight_Level_I': 2, 
                                        'Overweight_Level_II': 3, 
                                        'Obesity_Type_I': 4, 
                                        'Obesity_Type_II': 5, 
                                        'Obesity_Type_III': 6})
print(data['Obesity'].value_counts(normalize = True))

# Changing the column type to category:
for i in cat_columns:
    data[i] = data[i].astype('category')
```

```python
# separating x(features) and y(target):
y = data['Obesity']
x = data.drop(['Obesity'], axis = 1)
print(y.name)
print('')
print(x.columns)
```

```python
# Scaling the features in x:
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
x = pd.DataFrame(x_scaled, columns = x.columns)
```

```python
# Correlation Heatmap:
correlation_matrix = data.corr()
plt.figure(figsize = (10, 5))
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm', fmt = ".2f", cbar = True)

plt.title('Correlation Heatmap')
plt.ylabel('Features')
plt.show()
plt.clf()
```

MACHINE LERANING MODELS

```python
from sklearn.model_selection import train_test_split as tts

# Separate the dataset in train_dataset and test_dataset:
x_train, x_test, y_train, y_test = tts(x,y, test_size = 0.2, random_state = 42)
```

```python
# Using Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

model_rfc = RandomForestClassifier(n_estimators=100, random_state=42)

model_rfc.fit(x_train, y_train)
y_pred = model_rfc.predict(x_test)
```

```python
# Evaluating the model
from sklearn.metrics import accuracy_score, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Get feature importance scores
importances = model_rfc.feature_importances_
feature_names = x.columns

# Create a DataFrame
feat_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feat_importance_df = feat_importance_df.sort_values(by='Importance', ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 5))
plt.barh(feat_importance_df['Feature'], feat_importance_df['Importance'], color='skyblue')
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.title("Feature Importance - Random Forest")
plt.gca().invert_yaxis()
plt.show()

print(feat_importance_df)
```

```python
# Select only important features
sel_features = feat_importance_df[feat_importance_df['Importance'] > 0.03]['Feature']
x_train_sel = x_train.loc[:, sel_features]
x_test_sel = x_test.loc[:, sel_features]

model_rfc.fit(x_train_sel, y_train)
y_pred_sel = model_rfc.predict(x_test_sel)

print("Accuracy:", accuracy_score(y_test, y_pred_sel))
```

```python
# using the XGBoost:
import xgboost as xgb

model_xgb = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')

model_xgb.fit(x_train, y_train)
y_pred_xgb = model_xgb.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("\nClassification Report:\n", classification_report(y_test, y_pred_xgb))
```

```python
# Get feature importance scores
importances_xgb = model_xgb.feature_importances_

# Create a DataFrame
feat_importance_xgb = pd.DataFrame({'Feature': feature_names, 'Importance': importances_xgb})
feat_importance_xgb = feat_importance_xgb.sort_values(by='Importance', ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 5))
plt.barh(feat_importance_xgb['Feature'], feat_importance_xgb['Importance'], color='skyblue')
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.title("Feature Importance - XGBoost")
plt.gca().invert_yaxis()
plt.show()

print(feat_importance_xgb)
```

```python
# Select only important features
sel_features_xgb = feat_importance_xgb[feat_importance_xgb['Importance'] > 0.03]['Feature']
x_train_sel_xgb = x_train.loc[:, sel_features_xgb]
x_test_sel_xgb = x_test.loc[:, sel_features_xgb]

model_xgb.fit(x_train_sel_xgb, y_train)
y_pred_sel_xgb = model_xgb.predict(x_test_sel_xgb)

print("Accuracy:", accuracy_score(y_test, y_pred_sel_xgb))
```
